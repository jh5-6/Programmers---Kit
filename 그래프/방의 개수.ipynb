
# [방의 개수]

원점(0,0)에서 시작해서 아래처럼 숫자가 적힌 방향으로 이동하며 선을 긋습니다.



ex) 1일때는 오른쪽 위로 이동

그림을 그릴 때, 사방이 막히면 방하나로 샙니다.
이동하는 방향이 담긴 배열 arrows가 매개변수로 주어질 때, 방의 갯수를 return 하도록 solution 함수를 작성하세요.

< 제한사항 >
- 배열 arrows의 크기는 1 이상 100,000 이하 입니다.
- arrows의 원소는 0 이상 7 이하 입니다.
- 방은 다른 방으로 둘러 싸여질 수 있습니다.


[TIP] 

< 방의 생성 조건 >
    
    1. 방문한 노드가 이미 방문한 적이 있고,
    2. 해당 노드로 들어온 경로는 처음인 경우 
        
    * 예외 처리 : 모래시계 형태
        -> 한 번 이동할 때 두 칸씩 이동하도록 하여 해당 예외를 처리한다. 
        

Solution 

```python
from collections import deque 
from collections import defaultdict 

def solution(arrows):
    answer = 0
    
    dx = [0, 1, 1, 1, 0, -1, -1, -1]
    dy = [-1, -1, 0, 1, 1, 1, 0, -1]
    
    now = (0, 0)
    que = deque([now])
    
    
    # 노드 방문 여부 확인 & 경로 방문 확인
    visited_node = defaultdict(int)
    visited_edge = defaultdict(int)
    
    for arrow in arrows:
        
        # 모래시계 형태 처리를 위한 반복문
        for i in range(2):
            next = (now[0] + dy[arrow], now[1] + dx[arrow])
            que.append(next)
            now = next
            
    now = que.popleft() 
    visited_node[now] = 1
    
    while que:
        next = que.popleft()
        
        # next 노드를 이미 방문한 적이 있는 경우 
        if visited_node[next] == 1:
            
            # 해당 경로 처음 이용한 경우
            if visited_edge[(now, next)] == 0:
                answer += 1 
                
        else : 
            visited_node[next] = 1 
            
        visited_edge[(now, next)] = 1 
        visited_edge[(next, now)] = 1

        now = next
        
    return answer
```

```python
solution([6, 6, 6, 4, 4, 4, 2, 2, 2, 0, 0, 0, 1, 6, 5, 5, 3, 6, 0])
```
