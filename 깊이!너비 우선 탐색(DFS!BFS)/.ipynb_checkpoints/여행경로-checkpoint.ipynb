# [여행경로]

주어진 항공권을 모두 이용하여 여행경로를 짜려고 합니다. 항상 "ICN" 공항에서 출발합니다.

항공권 정보가 담긴 2차원 배열 tickets가 매개변수로 주어질 때, 방문하는 공항 경로를 배열에 담아 return 하도록 solution 함수를 작성해주세요.


    <제한사항>
    - 모든 공항은 알파벳 대문자 3글자로 이루어집니다.
    - 주어진 공항 수는 3개 이상 10,000개 이하입니다.
    - tickets의 각 행 [a, b]는 a 공항에서 b 공항으로 가는 항공권이 있다는 의미입니다.
    - 주어진 항공권은 모두 사용해야 합니다.
    - 만일 가능한 경로가 2개 이상일 경우 알파벳 순서가 앞서는 경로를 return 합니다.
    - 모든 도시를 방문할 수 없는 경우는 주어지지 않습니다.

    [TIP]
    
    * Test case 
    tickets = [['ICN','A], ['A', 'B'], ['A', 'C'], ['C', 'A'], ['B', 'D']] 인 경우






위와 같이 두가지 경로 가능하지만 B를 먼저갈 경우 모든 티켓을 이용할 수 없기 때문에 답이 될 수 없다. 

D까지 갔다가 길을 못찾고 A로 다시 돌아가야 하는데 어떻게 할까 고민하던 중 

길이 끊어진 D가 가장 마지막에 방문하는 공항이라고 생각해서 경로를 거꾸로 넣기로 했다.


1. 길이 끊어져 있는 answer에 D를 저장하고, D가없어지면 B도 끊어진다. 따라서 answer 에 B를 저장한다.


2. A는 B가 없어도 C가 있기 때문에 다시 진행한다. A -> C -> A -> B(X) -> D(X)


3. 가다보면 B는 이미 없어져 있으므로 A를 answer에 저장하고, 나머지 C, A 도 마찬가지로 저장한다.


4. 마지막 ICN 을 저장하면 ['D', 'B', 'A', 'C', 'A', 'ICN'] 이렇게 뒤집어진 경로가 나온다. 


5. 이 경로를 역순으로 출력해주면 된다.

<정확성 테스트 시간 초과>

```python
def solution(tickets):
    answer = []
    
    tickets.sort()
    
    start = "ICN"   
    while tickets :
        for i in range(len(tickets)):
            if tickets[i][0] == start:
                answer.append(start)
                start = tickets[i][1]
                tickets.remove(tickets[i])
       
                break
                
    answer.append(start)    
    return answer
```

[Solution] DFS 활용한 문제 풀이 

```python
from collections import defaultdict 

def solution(tickets):
    
    answer = [] 
    routes = defaultdict(list)
    
    for start, end in tickets:
        routes[start] = routes.get(start, []) + [end]
        
    for r in routes.keys():
        routes[r].sort(reverse = True)
        
    st = ["ICN"]
    while st:
        top = st[-1]
        
        if (not routes[top]) or (top not in routes):
            answer.append(st.pop())
        else :
            st.append(routes[top].pop())
            
    answer.reverse()
    return answer
```

```python
solution([["ICN", "SFO"], ["ICN", "ATL"], ["SFO", "ATL"], ["ATL", "ICN"], ["ATL","SFO"]])
```
